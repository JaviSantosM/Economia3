```python
#checar calculos mandados en whatsapp, utiles :)
def analisis_univariado(df_largo, indicador_analizar):
    """
    Realiza un análisis univariado detallado sobre una serie temporal
    a partir de un DataFrame en formato largo del Banco Mundial.

    Parámetros:
    - df_largo: DataFrame en formato long con columnas 'Series Name', 'Año' y 'Value'
    - indicador_analizar: str, nombre exacto del indicador a analizar

    Salida:
    - Imprime resumen .describe() y medidas personalizadas
    """

    print(f"\n--- Análisis Descriptivo para: {indicador_analizar} ---")

    # Filtrar los datos para la serie deseada
    serie = df_largo[df_largo["Series Name"] == indicador_analizar]["Value"].dropna()

    # Cálculo manual de medidas
    media = serie.mean()
    mediana = serie.median()
    desv_std = serie.std()
    varianza = serie.var()
    minimo = serie.min()
    maximo = serie.max()
    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)
    iqr = q3 - q1
    skewness = serie.skew()
    kurtosis = serie.kurt()

    # Resultados detallados
    print("\n Medidas Específicas:")
    print(f"Media: {media:.2f}")
    print(f"Mediana: {mediana:.2f}")
    print(f"Desviación Estándar (Volatilidad): {desv_std:.2f}")
    print(f"Varianza: {varianza:.2f}")
    print(f"Mínimo: {minimo:.2f}")
    print(f"Máximo: {maximo:.2f}")
    print(f"Rango (Max - Min): {(maximo - minimo):.2f}")
    print(f"Percentil 25 (Q1): {q1:.2f}")
    print(f"Percentil 75 (Q3): {q3:.2f}")
    print(f"Rango Intercuartílico (IQR): {iqr:.2f}")
    print(f"Asimetría (Skewness): {skewness:.2f}")
    print(f"Curtosis (Exceso): {kurtosis:.2f}")
```

```python
def formato_value(value):
  if pd.isna(value):
    return None
  if abs(value)>=1e12:
    return f'{value/1e12:.2f} T'
  if abs(value)>=1e9:
    return f'{value/1e9:.2f} B'
  if abs(value)>=1e6:
    return f'{value/1e6:.2f} M'
  if abs(value)>=1e3:
    return f'{value/1e3:.2f} k'
  if abs(value)<1e3:
    return f'{value}'
```

```python
df_pivot=df_subset_anim.pivot_table(
    index='Year',
    columns='Series Name',
    values='Value'
)
df_pivot.dropna(inplace=True)
df_pivot.reset_index(inplace=True)
df_pivot.head()
```

```python
all_frames_data=[]
for year_frame in df_pivot.Year:
  temp_df=df_pivot[df_pivot['Year']<=year_frame].copy()
  temp_df['FrameYear']=year_frame
  all_frames_data.append(temp_df)
all_frames_data
```

```python
df_cumulative=pd.concat(all_frames_data)
df_cumulative.head()
```

```python
df_cumulative['Point']=np.where(df_cumulative['Year']==df_cumulative['FrameYear'],15,5)
df_cumulative['Color']=np.where(df_cumulative['Year']==df_cumulative['FrameYear'],'Actual','Pasado')
df_cumulative.head()
```

```python
range_x_min = df_pivot[indicador_x].min() - abs(df_pivot[indicador_x].min() * 0.1) # Añadir margen
range_x_max = df_pivot[indicador_x].max() + abs(df_pivot[indicador_x].max() * 0.1)
range_y_min = df_pivot[indicador_y].min() - abs(df_pivot[indicador_y].min() * 0.1)
range_y_max = df_pivot[indicador_y].max() + abs(df_pivot[indicador_y].max() * 0.1)
```

```python
fig=px.scatter(df_cumulative,
              x=indicador_x,
              y=indicador_y,
              animation_frame='FrameYear',
              size='Point',
              color='Color',
              color_discrete_map={'Actual':'red','Pasado':'gray'},
              hover_name='Year',
              range_x=[range_x_min,range_x_max],
              range_y=[range_y_min,range_y_max]
)
fig.update_layout(transition= {'duration':100})
fig.update_layout(showlegend=False)
fig.show()
```
