```
def factorial(n):
  if(n==0 or n==1):
    return 1
  else:
    return n*factorial(n-1)
```

```
factorial(5)
```

```
def fibonacci(n):
  if(n==1):
    return 0
  elif(n==2):
    return 1
  else:
    return fibonacci(n-2)+fibonacci(n-1)
```

```
fibonacci(9)
```
