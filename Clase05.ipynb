```python
df_sp.colums=["Date","Close","High","Low","Open","Volume"]
df_sp.head()
```

```python
df_sp=df_sp.iloc[2:,:].reset_index(drop=True)
```

```python
df_sp.head()
```

```python
df_sp.info()
```

```python
df_sp["Price"]=pd.to_datetime(df_sp["Price"])
#df_sp["Price"]=df_sp.Price.astype("float")
df_sp["Close"]=df_sp.Close.astype("float")
df_sp["High"]=df_sp.High.astype("float")
df_sp["Low"]=df_sp.Low.astype("float")
df_sp["Open"]=df_sp.Open.astype("float")
df_sp["Volume"]=df_sp.Volume.astype("float")
df_sp.tail(10)
```

```python
df_sp[(df_sp.Open<250)]
```

```python
df_sp.loc[df_sp.Open<6800,"status"]="Atipico"
df_sp
```

```python
df_sp["Tendencia"]="Estable"
df_sp.head()
```

```python
df_sp.head()
```

```python
df_sp.loc[df_sp.Close>df_sp.Close.shift(1),"Tendencia"]="Sube"
df_sp.head(10)
```

```python
df_sp.loc[df_sp.Close<df_sp.Close.shift(1),"Tendencia"]="Baja"
df_sp.head(10)
```

```python
df_sp=df_sp.sort_values(by="Price")
df_sp.head(10)
```

```python
df_sp=df_sp.sort_values(by="High",ascending=False)
df_sp.head(10)
```
